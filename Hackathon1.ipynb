```python
data_train1=data_train.copy()
data_test1=data_test.copy()
```

```python
data_train1=data_train1.drop(labels=['risk_flag','Id'],axis=1)
data_test1=data_test1.drop(labels='id',axis=1)
```

```python
data_train1.sample(10)
```

```python
def cleaner(x):
  a= "".join(x.replace('_', ' ').split())

  return a
        
data_train1['city']=data_train1['city'].apply(cleaner)
data_train1['state']=data_train1['state'].apply(cleaner)
data_train1['profession']=data_train1['profession'].apply(cleaner)

data_test1['city']=data_test1['city'].apply(cleaner)
data_test1['state']=data_test1['state'].apply(cleaner)
data_test1['profession']=data_test1['profession'].apply(cleaner)
```

```python
data_train1.state.unique()
```

```python
data_test1.state.unique()
```

```python
data_train1.state.value_counts()
```

```python
data_unified=pd.concat([data_train1,data_test1],ignore_index=True)
data_unified
```

```python
data_unified['state'].value_counts()
```

```python
sensor_city =data_unified.city.value_counts().to_dict()
sensor_state =data_unified.state.value_counts().to_dict()
sensor_profession =data_unified.profession.value_counts().to_dict()

```

```python
data_unified['city']=data_unified['city'].map(sensor_city)
data_unified['state']=data_unified['state'].map(sensor_state)
data_unified['profession']=data_unified['profession'].map(sensor_profession)
```

```python
data_unified["profession"]=pd.factorize(data_unified.profession)[0]
data_unified["city"]=pd.factorize(data_unified.city)[0]
data_unified["state"]=pd.factorize(data_unified.state)[0]
data_unified["married"]=pd.factorize(data_unified.married)[0]
data_unified["house_ownership"]=pd.factorize(data_unified.house_ownership)[0]
data_unified["car_ownership"]=pd.factorize(data_unified.car_ownership)[0]
data_unified.sample(5)
```

```python
from sklearn.preprocessing import StandardScaler
s_scaler=StandardScaler()
col_names=list(data_unified.columns)
data_ss_unified=s_scaler.fit_transform(data_unified)
data_ss_unified=pd.DataFrame(data_ss_unified,columns=col_names)
data_ss_unified
```
